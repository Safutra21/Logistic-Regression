```python
X = df1[['Usia', 'Status', 'Kelamin', 'Memiliki_Mobil', 'Penghasilan']]
y = df1[['Beli_Mobil']]
```

```python
X
```

```python
y
```

```python
import sklearn.model_selection as ms

X_train, X_test, y_train, y_test = ms.train_test_split (X, y, test_size = 0.2, random_state =0)
print (X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
import sklearn.linear_model as lm

model =lm.LogisticRegression(solver='lbfgs')
model.fit(X_train, y_train)
print (model.coef_)
```

```python
y_prediksi = model.predict (X_test)
y_prediksi
```

```python
Prediksi = pd.DataFrame (y_prediksi, columns = ['Prediksi Beli Mobil'])
Prediksi
```

```python
import sklearn.metrics as met
confusionmatrix = met.confusion_matrix(y_test, y_prediksi)
print (confusionmatrix)
```

```python
from sklearn.metrics import classification_report


print(classification_report(y_test, y_prediksi))
```

```python
score = model.score (X_test, y_test)
print (score)
```

```python
score2 = met.accuracy_score(y_test,y_prediksi)
print (score2)
```

```python
precision = met.precision_score (y_test, y_prediksi)
print (precision)
```

```python
recall = met.recall_score (y_test, y_prediksi)
print (recall)
```

```python
import matplotlib.pyplot as plt

y_pred_proba =model.predict_proba (X_test) [::,1]
fp, tp, _ = met.roc_curve (y_test, y_pred_proba)
auc = met.roc_auc_score(y_test, y_prediksi)
print ("auc =", auc)
plt.plot (fp, tp, label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
import sklearn.feature_selection as fs

rfe = fs.RFE (model, 3)
rfe = rfe.fit (X_train, y_train)
print ('Support=', rfe.support_)
print ('Rangking=', rfe.ranking_)
```
